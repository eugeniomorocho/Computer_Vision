

2. **Importar las bibliotecas necesarias**: Primero, necesitamos importar las bibliotecas que vamos a utilizar. En este caso, necesitamos `cv2` para OpenCV y `matplotlib` para visualizar la imagen.



```python
import cv2
from matplotlib import pyplot as plt
```



3. **Leer una imagen**: Utilizamos la función `cv2.imread` de OpenCV para leer una imagen. La imagen debe estar en el mismo directorio que tu notebook de Jupyter, o necesitas proporcionar la ruta completa a la imagen.



```python
imgBGR = cv2.imread('/Users/eugenio/Documents/Notebooks_ArtificialIntelligence/5. Computer Vision/Image classification/gato.jpg')
```

4. **Visualizar la imagen**: OpenCV lee las imágenes en formato **BGR**, pero Matplotlib visualiza las imágenes en formato **RGB**. Por lo tanto, antes de visualizar la imagen, debemos convertirla de **BGR** a **RGB** usando el método `cv2.cvtColor` de la librería `opencv-python`. El argumento `cv2.COLOR_BGR2RGB` le dice a `cvtColor` qué conversión realizar: en este caso, de **BGR** a **RGB**.

```python
imgRGB = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB)
plt.imshow(imgRGB)
plt.show()
```
